### Cleaning Data


Membersihkan Tabel Geolocation


---



Pada hasil analisis sebelumnya pada tabel geolocation terdapat 261831 data duplikat yang dapat dihapus sebagai berikut:

```python
geolocation.drop_duplicates(inplace=True)

print("Data Duplikat geolocation :\n",geolocation.duplicated().sum(), "\n")
```

Membersihkan Tabel order_reviews

---



Pada Tabel order_reviews sebelumnya terdapat beberapa kesalahan yaitu terdapat beberapa data yang null. Sehingga dapat kita cek pada kolom review_comment_title, dan juga review_comment_message:

```python
order_reviews[order_reviews.review_comment_title.isna()]
```

**Insight:**
- Dapat dilihat dari hasil pengecekan pada kolom review_comment_title sebanyak 87656 data null dari 99223. Yang dimana hampir keseluruhan kolomnya null sehingga kita hapus saja untuk kolom review titlenya karena tidak terlalu penting.

```python
order_reviews.drop("review_comment_title", axis=1, inplace=True)
order_reviews.info()
```

Setelah itu, pada review_comment_message akan kita gunakan metode imputation dengan mengisikan kolom yang null dengan "no message"

```python
order_reviews["review_comment_message"].fillna("no message", inplace=True)
order_reviews.info()
```

```python
order_reviews.head()
```

**Insight:**
- Dapat dilihat sudah tidak terdapat nilai null yang ada pada tabel order_reviews. Selanjutnya yaitu mengganti tipe data dari review_creation_date, dan review_answer_timestamp menjadi tipe data datetime

```python
datetime_columns = ["review_creation_date", "review_answer_timestamp"]

for column in datetime_columns:
  order_reviews[column] = pd.to_datetime(order_reviews[column])

order_reviews.info()
```

membersihkan tabel orders

---

Sebelum membersihkan tabel orders, perlu dicatat berdasarkan pertanyaan bisnis yang telah dibuat, data yang akan digunakan hanya yang statusnya berhasil diantarkan sehingga kita akan memisahkan selain "delivered":

```python
print(orders["order_status"].value_counts())
```

```python
order_delivered = orders.drop(orders[orders['order_status'] != 'delivered'].index)

orders = order_delivered
orders.head()
```

```python
print(orders["order_status"].value_counts())
```

selanjutnya setelah memisahkan status order yang ada pada tabel orders, kita cek kembali value null yang ada pada tabel:

```python
print(orders.isna().sum())
```

dikarenakan jumlah null yang ada tidak terlalu banyak sehingga tidak terlalu mempengaruhi kualitas dataset, maka data yang null akan langsung dihapus:

```python
orders = orders.dropna(subset=["order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date"])
print(orders.isna().sum())
```

```python
orders.info()
```

Selanjutnya yaitu mengubah tipe data yang salah berdasarkan hasil analisis ke datetime:

```python
datetime_columns = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date"]

for column in datetime_columns:
  orders[column] = pd.to_datetime(orders[column])

orders.info()
```

selanjutnya tambahkan 1 kolom untuk mengecek apakah order diantar dengan tepat waktu atau terlambat:

```python
orders["deliver_time"] = np.where(orders["order_delivered_customer_date"] <= orders["order_estimated_delivery_date"], "On Time", "Late")

orders.sample(5)
```

Membersihkan Tabel order_items

---


Selanjutnya kita akan membersihkan tabel order_items, yang dimana berdasarkan hasil analisis sebelumnya terdapat kesalahan tipe data pada kolom shipping_limit_date yang akan kita ubah sebagai berikut

```python
order_items.info()
```

```python
datetime_columns = ["shipping_limit_date"]

for column in datetime_columns:
  order_items[column] = pd.to_datetime(order_items[column])

order_items.info()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Metode Pembayaran yang manakah paling banyak digunakan?

```python
data_order_payment = all_data.groupby(by="payment_type").size()
print(data_order_payment)
```

```python
plt.figure(figsize=(10, 6))
data_order_payment.plot(kind='bar', color='skyblue')
plt.title('Distribusi Metode Pembayaran')
plt.xlabel('Payment Type')
plt.ylabel('Number of Transactions')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```

**Insight:**
- Dapat dilihat pada hasil visualisasi diatas metode pembayaran yang terbanyak digunakan untuk melakukan transaksi yaitu credit_card sebanyak **85212**, dan disusul dengan boleto sebanyak **22441**.
